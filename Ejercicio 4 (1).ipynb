6. Demuestre que la escala integral de la función de correlación exponencial es igual al parámetro a y de la
función de correlación esférica es igual a (3/8)a.

La escala integral esta dada por:
$$I_z=\int_{0}^{\infty}\rho(\tau)d\tau$$

a) Exponencial 

$$C_z(h) = \sigma^2 \text{ exp}(-\frac{h}{30})$$ 

con $h>=0, a>0$, como $\rho(h) = \frac{C_z(h)}{\sigma^2}$ entonces se tiene que:

$$I_z = \int_{0}^{\infty}\text{ exp}(-\frac{h}{30})dh = -a \text{ exp}(-\frac{h}{30}) |_{0}^\infty = -a \text{ exp}(-\frac{\infty}{30})-(-a \text{ exp}(-\frac{\infty}{0}))=a   $$

Entonces $$I_z = a$$

Es decir a escala integral de la correlacion exponecial es $a$

b ) Esferica

$$C_z(h) = \sigma^2 \left\{
\begin{array}{ll}
1 - \frac{3}{2}\left(\frac{h}{a}\right) + \frac{1}{2}\left(\frac{h}{a}\right)^3, & \text{si } h < a \\
0, & \text{si } h \geq a
\end{array}
\right.
$$

omo $\rho(h) = \frac{C_z(h)}{\sigma^2}$ entonces se tiene que:

$$I_z=\int_{0}^{\infty}\frac{C_z(h)}{\sigma^2}dh =\int_{0}^{a} \left[1 - \frac{3}{2}\left(\frac{h}{a}\right) + \frac{1}{2}\left(\frac{h}{a}\right)^3\right] dh = \left[h - \frac{3}{4}\left(\frac{h^2}{a}\right) + \frac{1}{8}\left(\frac{h^4}{a^3}\right)\right]\bigg|_{0}^a = a - \frac{3}{4}a +\frac{1}{8}a = \frac{3}{8}a  $$ 

Entonces $$I_z = \frac{3}{8} a$$

7. Obtener la expresión:

$$C_z(\tau)=\int_{-\infty}^{\infty}S_z(\omega)cos(\omega\tau)d\omega$$

con función de densidad espectral:

$$S_z(\omega) = \frac{1}{2\pi}\int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\omega$$

Obtengamos primero la transformada de Fourier de la covarianza

$$\mathcal{F}\{C_z(\tau)\} = \int_{-\infty}^{\infty}C_z(\tau)e^{i\omega\tau}d\tau = \int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\tau+i\int_{-\infty}^{\infty}C_z(\tau)sin(\omega\tau)d\tau$$

Teniendo en cuenta que la funcion $C_z(\tau)$ es par debido a que $rho_z(\tau)$ es par y son proporcionales, se tiene que el termino :  $\int_{-\infty}^{\infty}C_z(\tau)sin(\omega\tau)d\tau=0$

Entonces la transformada de Fourier de la covarianza es:
$$\mathcal{F}\{C_z(\tau)\} = \int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\tau$$

Y es uan fuencion par, llamemosla 

$FC(\omega)= \int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\tau$

Ahora obtengamos su transformada inversa 

$$\mathcal{F}^{-1}\{\int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\tau\}=\mathcal{F}^{-1}\{FC(\omega)\} = \frac{1}{2\pi}\int_{-\infty}^{\infty}FC(\omega)e^{-i\omega\tau}d\omega=  \frac{1}{2\pi}\int_{-\infty}^{\infty}FC(\omega)cos(\omega\tau)d\omega-i\frac{1}{2\pi}\int_{-\infty}^{\infty}FC(\omega)sin(\omega\tau)d\omega=\frac{1}{2\pi}\int_{-\infty}^{\infty}FC(\omega)cos(\omega\tau)d\omega $$

Entonces la transformada inversa de $\mathcal{F}^{-1}\{FC(\omega)\} = \frac{1}{2\pi}\int_{-\infty}^{\infty}FC(\omega)cos(\omega\tau)d\omega $

Por tanto se tiene que:

$$\mathcal{F}^{-1}\{\mathcal{F}\{C_z(\tau)\}\} = \frac{1}{2\pi}\int_{-\infty}^{\infty}FC(\omega)cos(\omega\tau)d\omega= \frac{1}{2\pi}\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\tau cos(\omega\tau)d\omega$$

entonces $$C_z(\tau) = \mathcal{F}^{-1}\{\mathcal{F}\{C_z(\tau)\}\} = \int_{-\infty}^{\infty}FC(\omega)cos(\omega\tau)d\omega= \frac{1}{2\pi}\int_{-\infty}^{\infty}\bigg[\frac{1}{2\pi}\int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\tau \bigg]cos(\omega\tau)d\omega$$

Finalmente $$C_z(\tau)= \int_{-\infty}^{\infty}\bigg[\frac{1}{2\pi}\int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\tau \bigg]cos(\omega\tau)d\omega$$

Donde la densidad espectral esta dad por $Sz(\omega) = \frac{1}{2\pi}\int_{-\infty}^{\infty}C_z(\tau)cos(\omega\tau)d\tau$

Entonces $$C_z(\tau) = \frac{1}{2\pi}\int_{-\infty}^{\infty}Sz(\omega)cos(\omega\tau)d\omega$$

Y su par de fourier es $Sz(\omega)$

Ademas si evaluamos la covarianza en cero que nos da la varianza $C_z(0) = \sigma^2_z$, entonces 

$$C_z(0) = \frac{1}{2\pi}\int_{-\infty}^{\infty}Sz(\omega)cos(\omega*0)d\omega=  \frac{1}{2\pi}\int_{-\infty}^{\infty}Sz(\omega)d\omega $$

entonces 

$$C_z(0) =  \frac{1}{2\pi}\int_{-\infty}^{\infty}Sz(\omega)d\omega = \sigma_z^2$$
